Built in Django permissions

```python
from django.contrib.auth.models import Permission

# Super user = Admin user
# super_user = User.objects.filter(is_superuser=True).first()
# display(super_user.id)


# Staff user
staff_user = User.objects.filter(is_superuser=False, is_staff=True).first()
# display(staff_user.id)


# Check permissions in django
# permission_queryset = Permission.objects.all()
# for permission in permission_queryset:
#     print(permission.name)


# Giving Permissions to staff user
# permission = Permission.objects.filter(codename__icontains="view")
# for item in permission:
#     print(item.name)
#     staff_user.user_permissions.add(item)


# Giving Document model permissions to staff user
doc_permission_qs = Permission.objects.filter(codename__endswith="document")
for item in doc_permission_qs:
    print(item.name)
    staff_user.user_permissions.add(item)
```
